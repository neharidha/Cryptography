```python
#vigenere_encyption
def vigenere_enc(plaintext, keyword):
    ciphertext = ""
    key_index = 0
    plaintext=plaintext.replace(" ", '')
    plaintext=plaintext.upper()
    key=keyword.upper()
    for char in plaintext:
        if char.isalpha():
            key_char = key[key_index % len(key)].upper()
            key_value = ord(key_char) - ord('A')
            plaintext_index = ord(char.upper()) - ord('A')
            ciphertext_index = (plaintext_index + key_value) % 26
            ciphertext_char = chr(ciphertext_index + ord('A'))
            ciphertext += ciphertext_char
            key_index += 1
    return ciphertext


#vigenere_decyption
def vigenere_dec(ciphertext, keyword):
    plaintext = ""
    key_index = 0
    ciphertext=ciphertext.replace(" ", '')
    ciphertext=ciphertext.upper()
    key=keyword.upper()
    for char in ciphertext:
        if char.isalpha():
            key_char = key[key_index % len(key)].upper()
            key_value = ord(key_char) - ord('A')
            ciphertext_index = ord(char.upper()) - ord('A')
            plaintext_index = (ciphertext_index - key_value) % 26
            plaintext_char = chr(plaintext_index + ord('A'))
            plaintext += plaintext_char
            key_index += 1
    return plaintext

if __name__ == "__main__":                        
    input_str = 'test string' 
    input_key = 'keyword'                    
    encstr = vigenere_enc(input_str,input_key)
    print(encstr)              
    decstr = vigenere_dec(encstr,input_key)
    print(decstr) 
```
