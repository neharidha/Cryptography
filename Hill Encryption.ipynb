```python
#Hill encryption function
def hill_enc(input_matrix, input_plaintext):
    plaintext = input_plaintext.replace(" ", '')
    plaintext=plaintext.upper()
    pad = len(plaintext) % 3
    if pad == 1:
        plaintext += "XX"
    elif pad == 2:
        plaintext += "X"
    plaintext_matrix = np.zeros((len(plaintext) // 3, 3), dtype=int)
    for i in range(len(plaintext)):
        row = i // 3
        col = i % 3
        plaintext_matrix[row, col] = ord(plaintext[i]) - ord('A')
    
    
    ciphertext_matrix = np.matmul(plaintext_matrix, input_matrix) % 26
    ciphertext = ""
    for i in range(ciphertext_matrix.shape[0]):
        for j in range(ciphertext_matrix.shape[1]):
            ciphertext += chr(ciphertext_matrix[i, j] + ord('A'))
    return(ciphertext)

if __name__ == "__main__":                        
    input_matrix = np.array([[17,17,5],[21,18,21],[2,2,19]]) 
    input_plaintext = 'BCD'                    
    ciphertext = hill_enc(input_matrix,input_plaintext)
    print(ciphertext)
```
